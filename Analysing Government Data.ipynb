I successfully scraped a bunch of government data - now I can take a look at what I managed to scrape.

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn; seaborn.set()
```

```python
govt_data = pd.read_json('govtdata.json')
govt_data = govt_data[['title', 'agency', 'description', 'formats', 'link']]
```

```python
def grepl(pattern, col):
    return col.str.contains(pattern)
```

```python
def search_data(pattern, colname, search_in=""):
    if search_in == "": search_in = colname
    return govt_data.ix[grepl(pattern, govt_data[search_in].fillna("")), colname]
```

```python
def print_col(col):
    print("\n\n".join(sorted(["-%s" %s for s in list(col)])))
```

```python
print_col(search_data("Ministry of Health", 'description', search_in='agency'))
```

```python
print_col(search_data("Statistics New Zealand", "description", search_in='agency'))
```

## Creating a word cloud

Now I can have a go at visualising the text from the descriptions of the data. This should be a way of informing myself about which agencies are worth looking into.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
desc_words = '\n'.join(govt_data.description.apply(lambda x: x if x is not None else ''))
```

```python
print(desc_words)
```

```python
agencies = np.unique(govt_data.agency)
```

```python
for agency in agencies:
    ag_words = '\n'.join(govt_data[govt_data.agency == agency].description.apply(lambda x: x if x is not None else ''))
    try:
        wordcloud = WordCloud(stopwords=STOPWORDS,
                              background_color='white',
                              width=1200,
                              height=1000
                             )
        wordcloud = wordcloud.generate(ag_words)
    except ZeroDivisionError:
        print("Skipped '%s'" % agency)
        continue
    print(agency)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.savefig('wordclouds/%s.png' % agency)
```
