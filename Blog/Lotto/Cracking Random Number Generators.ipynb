# Cracking Random Number Generators

Here I'll be attempting to crack the python random number generator.

```python
import random
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
shape = (10**4, 10**4)
X = np.random.random(shape)
plt.imshow(X, cmap='hot', interpolation='nearest')
```

Looks pretty random to me.

Maybe we need to look at it in a more 'sequential' way.

```python
random.seed(1)
x_1 = random.random()
x_1
```

```python
for i in range(1, 10**11+1):
    if i % 10**6 == 0:
        print("\riteration %d" %i, end="")
    if random.random() == x_1:
        print(i)
```
